```python
import cv2

cap=cv2.VideoCapture(0)

#Automatically grab width and height from video feed
#(returns float which we need to convert to integer foer later on)

width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
#we're using here because in python alllowa=s for int classical division,
# because we can't pass a float to the cv2.rectangle fuction

# Coordinates for rectangle, center point rectangle
x=width//2
y=height//2

#width and height adding 1/4 of width and height from the center
w=width//4
h=height//4

while True:
    #Capture frame-by-frame
    ret, frame = cap.read()
    
    #Draw a rectangle on stream
    cv2.rectangle(frame,(x,y),(x+w, y+h), color=(0,0,255), thickness=4)
    
    #display the resulting frame
    cv2.imshow('frame', frame)
    
    #ths command let's us quit with the"q" button on a keyboard
    #simply pressing x on window won't work
    
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        break
        
#when everything is done release the capture
cap.release()
cv2.destroyAllWindows()
```
